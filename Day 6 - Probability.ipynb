# Probability

Probablility = desired number / all possibility 

# Conditional Probablility

```python
spam = {
    "password": 2,
    "review": 1,
    "send": 3,
    "us": 3,
    "your": 3,
    "account": 1
}

p_password_given_spam = spam['password']/sum(spam.values())
print(p_password_given_spam)

ham = {
    "password": 1,
    "review": 2,
    "send": 1,
    "us": 1,
    "your": 2,
    "account": 0
}

p_password_given_ham = ham['password']/sum(ham.values())
print(p_password_given_ham)
```

# Probability Function!!!

```python
def probability (data, key):
    password_count = data[key]
    total = sum(data.values())
    probability = password_count / total
    return probability

print(probability(spam, 'password'))
print(probability(ham, 'password'))
```
